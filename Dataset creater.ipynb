2. Download tiger subspecies images

```python
# ============================================
# CELL 2: Download Tiger Images Using BING (Working!)
# ============================================

# Create download directory
DOWNLOAD_DIR = Path.cwd() / "tiger_dataset"
DOWNLOAD_DIR.mkdir(exist_ok=True)

# Tiger subspecies
SUBSPECIES = ["bengal tiger", "siberian tiger", "malayan tiger"]
IMAGES_PER_CLASS = 100

print("="*60)
print("DOWNLOADING TIGER IMAGES FROM BING")
print("="*60)
print(f"Save location: {DOWNLOAD_DIR.absolute()}")
print(f"Images per class: {IMAGES_PER_CLASS}")
print("="*60)

for species in SUBSPECIES:
    print(f"\n📥 Downloading {species}...")
    
    folder_name = species.replace(" ", "_")
    save_path = DOWNLOAD_DIR / folder_name
    save_path.mkdir(exist_ok=True)
    
    # Bing crawler - this works reliably
    bing_crawler = BingImageCrawler(
        storage={'root_dir': str(save_path)},
        downloader_threads=4,
        feeder_threads=1,
        parser_threads=1
    )
    
    # Optional filters for better quality
    filters = dict(
        size='large',      # Large images
        type='photo',      # Real photos only
        color='color'      # Color images
    )
    
    try:
        bing_crawler.crawl(
            keyword=species,
            filters=filters,
            max_num=IMAGES_PER_CLASS,
            min_size=(300, 300),
            file_idx_offset=0
        )
        print(f"✅ Completed {species}")
    except Exception as e:
        print(f"❌ Error: {e}")

print("\n" + "="*60)
print("🎉 DOWNLOAD COMPLETE!")
print(f"📁 Files saved to: {DOWNLOAD_DIR.absolute()}")
print("="*60)
```

5. Split the dataset to train/validation/test

```python
# ============================================
# CELL 5: Standardize Images (Resize to 224x224)
# ============================================

def standardize_images(input_dir, output_dir, image_size=(224, 224)):
    """
    Standardize images: convert to RGB, resize to target size, save as JPG
    """
    input_dir = Path(input_dir)
    output_dir = Path(output_dir)
    
    print("\n" + "="*60)
    print("STANDARDIZING IMAGES")
    print("="*60)
    print(f"Source: {input_dir}")
    print(f"Target: {output_dir}")
    print(f"Target size: {image_size[0]}x{image_size[1]}")
    print("-"*60)
    
    stats = {}
    total_processed = 0
    
    for class_folder in input_dir.iterdir():
        if not class_folder.is_dir():
            continue
        
        class_name = class_folder.name
        stats[class_name] = {'success': 0, 'failed': 0}
        
        print(f"\n📁 Processing: {class_name}")
        
        # Find all images
        image_files = []
        for ext in ['*.jpg', '*.jpeg', '*.png', '*.webp', '*.bmp']:
            image_files.extend(class_folder.glob(ext))
        
        print(f"   Found {len(image_files)} images")
        
        for img_path in image_files:
            # Create save path
            relative_path = img_path.relative_to(input_dir)
            save_path = output_dir / relative_path.with_suffix('.jpg')
            save_path.parent.mkdir(parents=True, exist_ok=True)
            
            try:
                # Open, convert to RGB, resize, save
                with Image.open(img_path) as img:
                    img_rgb = img.convert('RGB')
                    img_resized = img_rgb.resize(image_size, Image.Resampling.LANCZOS)
                    img_resized.save(save_path, 'JPEG', quality=95, optimize=True)
                    stats[class_name]['success'] += 1
                    total_processed += 1
                    
            except Exception as e:
                stats[class_name]['failed'] += 1
                print(f"   ⚠️ Failed: {img_path.name} - {str(e)[:50]}")
        
        print(f"   ✅ Success: {stats[class_name]['success']}, Failed: {stats[class_name]['failed']}")
    
    print("\n" + "-"*60)
    print(f"📊 TOTAL STANDARDIZED: {total_processed} images")
    print("✅ Standardization complete!")
    
    return stats

# Standardize images
STANDARDIZED_DIR = Path.cwd() / "standardized_dataset"
standardization_stats = standardize_images(DOWNLOAD_DIR, STANDARDIZED_DIR, (224, 224))
```

6. final dataset summary

```python
# ============================================
# CELL 6: Split Dataset (Train/Validation/Test)
# ============================================

def split_dataset(source_dir, output_dir, train_ratio=0.70, val_ratio=0.15, seed=42):
    """
    Split dataset into train (70%), validation (15%), test (15%)
    """
    random.seed(seed)
    source_dir = Path(source_dir)
    output_dir = Path(output_dir)
    
    print("\n" + "="*60)
    print("SPLITTING DATASET")
    print("="*60)
    print(f"Split ratio: Train={train_ratio*100:.0f}%, Validation={val_ratio*100:.0f}%, Test={100-(train_ratio+val_ratio)*100:.0f}%")
    print("-"*60)
    
    split_stats = {}
    
    for class_folder in sorted(source_dir.iterdir()):
        if not class_folder.is_dir():
            continue
        
        # Get all standardized JPG images
        images = list(class_folder.glob("*.jpg"))
        random.shuffle(images)
        
        total = len(images)
        train_end = int(total * train_ratio)
        val_end = train_end + int(total * val_ratio)
        
        splits = {
            "train": images[:train_end],
            "validation": images[train_end:val_end],
            "test": images[val_end:]
        }
        
        split_stats[class_folder.name] = {
            "train": len(splits["train"]),
            "validation": len(splits["validation"]),
            "test": len(splits["test"]),
            "total": total
        }
        
        # Copy files to split folders
        for split_name, files in splits.items():
            target_folder = output_dir / split_name / class_folder.name
            target_folder.mkdir(parents=True, exist_ok=True)
            
            for file in files:
                shutil.copy2(file, target_folder / file.name)
        
        print(f"\n{class_folder.name}:")
        print(f"   Train: {len(splits['train'])}")
        print(f"   Validation: {len(splits['validation'])}")
        print(f"   Test: {len(splits['test'])}")
        print(f"   Total: {total}")
    
    print("\n" + "-"*60)
    print("✅ Dataset split complete!")
    
    return split_stats

# Split the dataset
FINAL_DATASET_DIR = Path.cwd() / "animal_subspecies_dataset"
split_stats = split_dataset(STANDARDIZED_DIR, FINAL_DATASET_DIR)
```

7. Check for corrupted images

```python
# ============================================
# CELL 7: Final Dataset Summary
# ============================================

def create_final_summary(dataset_dir):
    """Create comprehensive summary of final dataset"""
    dataset_dir = Path(dataset_dir)
    
    print("\n" + "="*60)
    print("FINAL DATASET SUMMARY")
    print("="*60)
    
    summary_data = []
    total_per_split = {}
    
    for split in ["train", "validation", "test"]:
        split_dir = dataset_dir / split
        if split_dir.exists():
            split_total = 0
            for class_dir in sorted(split_dir.iterdir()):
                if class_dir.is_dir():
                    count = len(list(class_dir.glob("*.jpg")))
                    split_total += count
                    summary_data.append({
                        "Split": split.capitalize(),
                        "Class": class_dir.name,
                        "Images": count
                    })
            total_per_split[split] = split_total
    
    # Display as table
    if summary_data:
        import pandas as pd
        df = pd.DataFrame(summary_data)
        pivot_df = df.pivot(index="Class", columns="Split", values="Images").fillna(0).astype(int)
        pivot_df["Total"] = pivot_df.sum(axis=1)
        print("\n" + pivot_df.to_string())
        
        print("\n" + "-"*60)
        print("OVERALL STATISTICS:")
        print(f"   Total images: {sum(total_per_split.values())}")
        print(f"   Training: {total_per_split.get('train', 0)}")
        print(f"   Validation: {total_per_split.get('validation', 0)}")
        print(f"   Testing: {total_per_split.get('test', 0)}")
        print(f"   Number of classes: {len(pivot_df)}")
        print(f"   Image size: 224x224x3 (RGB)")
        print(f"   Split ratio: 70/15/15")
        
        return total_per_split
    else:
        print("❌ No data found in final dataset directory")
        return None

# Create final summary
final_stats = create_final_summary(FINAL_DATASET_DIR)
```

8. Remove corrupted images

```python
# ============================================
# CELL 8: Check for Corrupted Images
# ============================================

def check_corrupted_images(dataset_dir):
    """Check for and report corrupted images"""
    dataset_dir = Path(dataset_dir)
    
    print("\n" + "="*60)
    print("CHECKING FOR CORRUPTED IMAGES")
    print("="*60)
    
    corrupted = []
    total_checked = 0
    
    for image_path in dataset_dir.rglob("*.jpg"):
        total_checked += 1
        try:
            with Image.open(image_path) as img:
                img.verify()
        except Exception as e:
            corrupted.append(str(image_path))
            print(f"⚠️ Corrupted: {image_path.relative_to(dataset_dir)}")
    
    print("-"*60)
    print(f"📊 Total images checked: {total_checked}")
    print(f"✅ Corrupted images found: {len(corrupted)}")
    
    return corrupted

# Check for corrupted images
corrupted_images = check_corrupted_images(FINAL_DATASET_DIR)

if corrupted_images:
    print("\n🗑️ Removing corrupted images...")
    for img_path in corrupted_images:
        try:
            Path(img_path).unlink()
            print(f"   Removed: {Path(img_path).name}")
        except:
            print(f"   Failed: {Path(img_path).name}")
    print("✅ Corrupted images removed!")
```
